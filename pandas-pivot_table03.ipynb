```python
df2 = pd.DataFrame({"A": ["foo", "foo", "foo", "foo", "foo",
                         "bar", "bar", "bar", "bar"],
                   "B": ["one", "one", "one", "two", "two",
                          "one", "one", "two", "two"],
                    "C": ["small", "large", "large", "small",
                          "small", "large", "small", "small",
                      "large"],
                    "D": [1, 2, 2, 3, 3, 4, 5, 6, 7]})
```

```python
df2
```

```python
table1 = pd.pivot_table(df2,values='D',index=['A','B'],columns=['C'],aggfunc=np.sum)
```

```python
table1.to_excel('./tables01.xlsx',engine='openpyxl')
```

```python
# 查询业绩数据 2017/11/29
from sqlalchemy import create_engine
engine = create_engine('mysql+pymysql://uapi:G#!2017.data@127.0.0.1:3306/DbReport?charset=utf8')
df3 = pd.read_sql('select * from F_ErpClassStat_Financial where reck_date="2017.11.01"',engine)
```

```python
df3.head()
```

```python
# 使用pivot_table  透视表,实现项目 行转列
#变量“columns（列）”是可选的，它提供一种额外的方法来分割你所关心的实际值。
#然而，聚合函数aggfunc最后是被应用到了变量“values”中你所列举的项目上。
#非数值（NaN）有点令人分心。如果想移除它们，我们可以使用“fill_value”将其设置为0。
#如果将参数margins设置为True，则可以得到分项总计数据。
table2 = pd.pivot_table(df3[['City_Id','Shop_Name','Dept_Name','Doctor_Name','Third_Name','Total_Mount','Pay_Money']],
                           values='Pay_Money',
                           index=['City_Id','Shop_Name','Dept_Name','Doctor_Name'],
                           columns=['Third_Name'],
                           aggfunc=np.sum,
                           #margins=True,
                           fill_value=0)
```

```python
table2.head()
```

```python
table2.to_excel('text_pandaexcel03.xlsx',engine='openpyxl')
```

```python
table3 = pd.pivot_table(df3[['City_Id','Shop_Name','Dept_Name','Doctor_Name','Third_Name','Total_Mount','Pay_Money']],
                       values=['Total_Mount','Pay_Money'],
                       index=['City_Id','Shop_Name','Dept_Name','Doctor_Name','Third_Name'],
                       aggfunc=np.sum,
                       margins=True,
                       fill_value=0,
                       dropna = True
                       )
```
